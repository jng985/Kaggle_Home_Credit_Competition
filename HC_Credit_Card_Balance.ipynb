```python
from IPython.core.display import display, HTML
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
import helper_functions as hf

warnings.filterwarnings('ignore')
```

```python
cc = pd.read_csv('credit_card_balance.csv').sort_values(by=['SK_ID_PREV','MONTHS_BALANCE'])
cc['NAME_CONTRACT_STATUS'] = [1 if i=='Active' else 0 for i in cc['NAME_CONTRACT_STATUS']]
cc['AMT_BALANCE'] = cc['AMT_BALANCE'].replace(0,np.nan)
cc.head()
```

```python
cc['cc_f1'] = cc['AMT_CREDIT_LIMIT_ACTUAL']/cc['AMT_BALANCE']
cc['cc_f2'] = cc['AMT_DRAWINGS_ATM_CURRENT']/cc['AMT_BALANCE']
cc['cc_f3'] = cc['AMT_INST_MIN_REGULARITY'] / cc['AMT_PAYMENT_CURRENT']
cc['cc_f4'] = cc['AMT_PAYMENT_TOTAL_CURRENT'] / cc['AMT_CREDIT_LIMIT_ACTUAL']
cc['cc_f5'] = cc['AMT_RECIVABLE'] / cc['AMT_RECEIVABLE_PRINCIPAL']
cc['cc_f6'] = cc['AMT_TOTAL_RECEIVABLE'] / cc['AMT_RECEIVABLE_PRINCIPAL']
cc['cc_f7'] = hf.cnt_percents(cc,'CNT_DRAWINGS_ATM_CURRENT','CNT_DRAWINGS_CURRENT')
cc['cc_f8'] = hf.cnt_percents(cc,'CNT_DRAWINGS_OTHER_CURRENT','CNT_DRAWINGS_CURRENT')
cc['cc_f9'] = hf.cnt_percents(cc,'CNT_DRAWINGS_POS_CURRENT','CNT_DRAWINGS_CURRENT')
```

```python
cc_prev_aggs_dict = {'MONTHS_BALANCE':['count','mean']}
for i in cc.columns[3:]:
    if i.startswith('AMT'):
        cc_prev_aggs_dict[i] = ['min','max','mean','sum']
    if i.startswith('CNT'):
        cc_prev_aggs_dict[i] = ['min','max','mean','var','sum']
    if i.startswith('SK_DPD'):
        cc_prev_aggs_dict[i] = ['max','mean','var','sum']
    if i.startswith('cc_f'):
        cc_prev_aggs_dict[i] = ['min','max','mean','var','nunique']
cc_prev_aggs = hf.get_aggs_df(cc,'SK_ID_PREV',cc_prev_aggs_dict)
```

```python
cc_prev_aggs_dict
```

```python
cc_prev_slopes = hf.get_slopes(cc,cc['SK_ID_PREV'].unique(),'SK_ID_PREV','MONTHS_BALANCE',cc.columns[3:])
cc_prev_slopes.to_csv('cc_prev_slopes.csv',columns = cc_prev_slopes.columns)
cc_prev_slopes = pd.read_csv('cc_prev_slopes.csv').set_index('SK_ID_PREV')
cc_prev_slopes.head()
```

```python
cc_prev_weighted = hf.get_weighted(cc,'SK_ID_PREV','MONTHS_BALANCE',cc.columns[3:])
cc_prev_weighted.to_csv('cc_prev_weighted.csv',columns = cc_prev_weighted.columns)
cc_prev_weighted = pd.read_csv('cc_prev_weighted.csv').set_index('SK_ID_PREV')
cc_prev_weighted.head()
```

```python
cc_prev = pd.concat([cc_prev_aggs, cc_prev_slopes,cc_prev_weighted],axis=1)
cc_prev['SK_ID_CURR'] = cc.groupby('SK_ID_PREV').max()['SK_ID_CURR']
cc_prev.head()
```

```python
cc_curr = cc_prev.groupby('SK_ID_CURR').agg(['min','max','mean'])
cc_curr.columns = hf.join_tuples_cols(cc_curr)
cc_curr = hf.drop_high_corrs(cc_curr,0.95)
cc_curr.columns = ['cc_'+i for i in cc_curr.columns]
cc_curr = cc_curr.replace([-np.inf,np.inf],np.nan)
cc_curr.to_csv('cc_curr.csv',columns = cc_curr.columns)
cc_curr = pd.read_csv('cc_curr.csv').set_index('SK_ID_CURR')
cc_curr.head()
```
